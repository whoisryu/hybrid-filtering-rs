```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_recall = pd.DataFrame({
    'Name': ['Word2Vec','RBM','Hybird'],
    'Values': [0.0,0.0,0.02],
})


df_precission = pd.DataFrame({
     'Name': ['Word2Vec','RBM','Hybird'],
     'Values': [0.20,0.20,0.10]
})
```

```python
ax = df_recall.plot(x='Name', kind='bar', title='Recall@5')

ax.set_ylabel('Values')
```

```python
ax = df_precission.plot(x='Name', kind='bar', title='Precission@5')

ax.set_ylabel('Values')
```
